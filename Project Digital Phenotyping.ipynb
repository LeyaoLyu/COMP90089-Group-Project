# Diagnostic criteria:
If any one of the three criteria are fulfilled:

Criterion 1:
* Acute onset of illness within minutes or hours involving skin and/or mucosal tissue(lips, mouth, tongue etc.)[skin symptoms present in 90% of cases and at least one of:
  * Respiratory compromise
  * Reduced blood pressure or signs of end-organ mal-perfusion

Criterion 2:
* After exposure to a likely allergen (minutes to several hours), patient experiencing 2 or more of the below
  * Involvement of Skin-mucosal tissue
  * Respiratory compromise
  * Reduced Blood pressure or signs of end-organ malfusion
  * Persistent gastrointestinal symptoms

Criterion 3:
Reduced Blood Pressure (systolic BP of **less than 90mmHg** or 30% decrease from patient's baseline) after exposure to known allergen

Possible organ effects:
  * Cutaneous system (eg, urticaria, angioedema, pruritus)
  * Respiratory system (eg, dyspnea, cough, wheezing)
  * Cardiovascular system (eg, hypotension, angina, arrhythmias)
  * Gastrointestinal system (eg, crampy abdominal pain, vomiting)
  * Central nervous system (eg, confusion, dizziness, irritability, headache)

# Treatment:
* Intramuscular or Intravenous injection of epinephrine (adrenaline) (intravenous if severe)
  * if patient taking beta blockers may need to supplement with glucagon (1 to 5mg slow IV bolus over 5 minutes, can be followed by titrated infusion of 5 to 15 mcg/minute)
* Patient placement (doubt this is useful for us)
* Supplemented oxygen (nonrebreather mask at 15liters/minute or commerical hih oxygen flow mask, providing at least 70% to 100% oxygen)
* IV Fluids
  * isotonic (0.9%) saline to mainatin venous access
  * Possible fluid resuscitation using normla saline and rapid flow and 1-2L of fluid in he first minutes, up to 7 ltres total may be used
* Continuous monitioring of Blood Pressure, heart rate, respiratory rate and oxygen saturation
* Airway
  * Inubation if respiratory distress
  * Early preparaion for intubation if airway involvement or edema(swelling due to fluid) of upper airway
  * can require emergency cricothyroidotomy
